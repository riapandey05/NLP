```python
import pandas as pd
messages=pd.read_csv('spam.csv',sep=',',names=["label","message"],encoding='ISO-8859-1',usecols=[0, 1])
```

```python
import re
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-z]',' ',str(messages['message'][i]))
    review = review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review ]
    review=' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus,messages['message']) if i<1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[] 
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
#simple_preprocess(sent) is a function from the Gensim library that performs the following:
# Tokenizes the sentence into words.
# Converts words to lowercase.
# Removes any non-alphabetic characters (so only words made up of a-z are retained).
# Optionally, it may also remove stopwords depending on the configuration.

```
